**Instructions:**

Design and implement an application that helps users create and manage a monthly
budget. The application should prompt the user to input their monthly income and
various expense categories (e.g., rent, groceries, utilities, entertainment).
Your design should:

1.  Prompt the user for their monthly income and amounts for each expense
    category.
2.  Calculate and display the total expenses and the remaining balance after all
    expenses.
3.  Allow the user to repeat the process with different inputs to plan different
    budget scenarios.

Your design should demonstrate the first five steps of the development

**[25 Marks]**

```python

income_prompt = "Enter your monthly income: $"
rent_prompt = "Enter your monthly rent: $"
groceries_prompt = "Enter your monthly groceries budget: $"
utilities_prompt = "Enter your monthly utilities budget: $"
entertainment_prompt = "Enter your monthly entertainment budget: $"

def get_user_input(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value < 0:
                print("Please enter a positive value.")
            else:
                return value
        except ValueError:
            print("Invalid input. Please enter a numeric value.")


def calculate_and_display_budget():
    monthly_income = get_user_input(income_prompt)
    monthly_rent = get_user_input(rent_prompt)
    monthly_groceries = get_user_input(groceries_prompt)
    monthly_utilities = get_user_input(utilities_prompt)
    monthly_entertainment = get_user_input(entertainment_prompt)

    total_expenses = monthly_rent + monthly_groceries + monthly_utilities + monthly_entertainment
    remaining_balance = monthly_income - total_expenses


    print("\nMonthly Budget Summary:")
    print("-" * 30)
    print(f"Income: ${monthly_income:.2f}")
    print("Expenses:")
    print(f"\tRent: ${monthly_rent:.2f}")
    print(f"\tGroceries: ${monthly_groceries:.2f}")
    print(f"\tUtilities: ${monthly_utilities:.2f}")
    print(f"\tEntertainment: ${monthly_entertainment:.2f}")
    print("-" * 30)
    print(f"Total Expenses: ${total_expenses:.2f}")
    print(f"Remaining Balance: ${remaining_balance:.2f}")

def main():
    while True:
        calculate_and_display_budget()
        repeat = input("\nDo you want to plan another budget scenario? (yes/no): ").strip().lower()
        if repeat != 'yes':
            break

if __name__ == "__main__":
    main()
```

```python
# First Level: The Neighbourhood Watch

def schedule_volunteers(volunteers: list, shifts: list) -> dict:
    """
    Assigns volunteers to available shifts based on their preferences.

    Each volunteer is asked to choose their preferred shift from the available options.
    The function then assigns the shifts based on the provided preferences.

    Args:
        volunteers (list): A list of volunteer names.
        shifts (list): A list of available shifts.

    Returns:
        dict: A dictionary mapping each shift to the assigned volunteer.

    Example:
        >>> schedule_volunteers(["Alice", "Bob"], ["Morning", "Afternoon"])
        {'Morning': 'Alice', 'Afternoon': 'Bob'}
    """
    assignments = {}

    for volunteer in volunteers:
        while True:
            try:
                print(f"Available shifts: {shifts}")
                preference_num = int(input(f"{volunteer}, enter your preferred shift number (1-{len(shifts)}): "))
                if preference_num < 1 or preference_num > len(shifts):
                    raise ValueError("Invalid shift number. Please try again.")
                break
            except ValueError as e:
                print(e)

        # Convert to 0-based index
        shift_index = preference_num - 1
        chosen_shift = shifts.pop(shift_index)
        assignments[chosen_shift] = volunteer

    return assignments

if __name__ == "__main__":
    # List of volunteers
    volunteers = ["Alice", "Bob"]
    # List of shifts
    available_shifts = ["Morning", "Afternoon"]

    print("Welcome to Code City Adventures: The Neighbourhood Watch")
    print("You will help organise a schedule for volunteer watch shifts.")

    result = schedule_volunteers(volunteers, available_shifts)

    print("\nVolunteer Assignments:")
    for shift, volunteer in result.items():
        print(f"{shift}: {volunteer}")
```
